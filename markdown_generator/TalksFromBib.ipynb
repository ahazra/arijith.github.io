```python
import bibtexparser
import pandas as pd
import os
```

```python
with open('Talks.bib') as bibtex_file:
    bib_database = bibtexparser.load(bibtex_file)
```

```python
talks = pd.DataFrame(bib_database.entries)

talks.sort_values(by='date', ascending = False, inplace = True)
```

```python
html_escape_table = {
    "&": "&amp;",
    '"': "&quot;",
    "'": "&apos;"
    }

def html_escape(text):
    if type(text) is str:
        return "".join(html_escape_table.get(c,c) for c in text)
    else:
        return "False"
```

```python
for row, item in talks.iterrows():
    
    md_filename = str(item.ID) + "-" + str(item.name+1) + ".md"
    html_filename = str(item.ID) + "-" + str(item.name+1) 
    year = item.date[:4]

    
    md = "---\ntitle: \""   + item.title + '"\n'
    md += "collection: talks" + "\n"   
    if len(str(item.type)) > 3:
        md += 'type: "' + item.type + '"\n'
    else:
        md += 'type: "Talk"\n'
    
    md += "permalink: /talks/" + html_filename + "\n"
    
    if len(str(item.note)) > 3:
        md += 'venue: "' + item.note + '"\n'
        
    if len(str(item.date)) > 3:
        md += "date: " + str(item.date) + "\n"
    
    if len(str(item.location)) > 3:
        md += 'location: "' + str(item.location) + '"\n'
           
    md += "---\n"
    
    

        
    talk_dir = os.path.abspath(os.path.join(os.pardir,'_talks'))
    
    if not os.path.exists(talk_dir):
        os.makedirs(talk_dir)
    
    f_name = os.path.join(talk_dir,md_filename)
    #print(md)
    print(f_name)

    with open(f_name, 'w') as f:
        f.write(md)
```

```python
for row, item in talks.iterrows():
    
    md_filename = str(item.id) + "-" + item.name + ".md"
    html_filename = str(item.date) + "-" + item.url_slug 
    year = item.date[:4]
    
    md = "---\ntitle: \""   + item.title + '"\n'
    md += "collection: talks" + "\n"   
    if len(str(item.type)) > 3:
        md += 'type: "' + item.type + '"\n'
    else:
        md += 'type: "Talk"\n'
    
    md += "permalink: /talks/" + html_filename + "\n"
    
    if len(str(item.venue)) > 3:
        md += 'venue: "' + item.venue + '"\n'
        
    if len(str(item.location)) > 3:
        md += "date: " + str(item.date) + "\n"
    
    if len(str(item.location)) > 3:
        md += 'location: "' + str(item.location) + '"\n'
           
    md += "---\n"
    
    
    if len(str(item.talk_url)) > 3:
        md += "\n[More information here](" + item.talk_url + ")\n" 
        
    
    if len(str(item.description)) > 3:
        md += "\n" + html_escape(item.description) + "\n"
        
    talk_dir = os.path.abspath(os.path.join(os.pardir,'_talks'))
    
    if not os.path.exists(talk_dir):
        os.makedirs(talk_dir)
    
    f_name = os.path.join(talk_dir,md_filename)
    #print(md)
    print(f_name)

    with open(f_name, 'w') as f:
        f.write(md)
```
